```python
import numpy as np
import math
import matplotlib.pyplot as plt
```

```python
def Q_rsqrt(number):
    threehalfs = 1.5
    x2 = number * 0.5
    y = np.float32(number)
    i = y.view(np.int32)
    i = np.int32(0x5f3759df) - (i >> 1)
    y = i.view(np.float32)
    y = y * (threehalfs - (x2 * y * y))
    return y
```

```python
def plot():
    x = np.arange(1, 100000, 0.1)
    y = [(1/math.sqrt(i) - Q_rsqrt(i)) for i in x]
    plt.plot(x, y)
    plt.show()
```

```python
def listing():
    for i in range(1, 10000):
        print(Q_rsqrt(i))
```

```python
if __name__ == "__main__":
    listing()
```

```python

if __name__ == '__main__':
    plot()
```
